# Classical Solver

```python
# imports

from __future__ import annotations

import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))

import matplotlib.pyplot as plt
from src.ref_solutions import PDESolver
```

```python
solver = PDESolver(dim=1, nx=100, Lx=1.0)
history = solver.solve(
    equation="burgers",
    nu=0.1,
    dt=0.01,
    steps=200,
    sigma=0.1
)

# history[k] = solution at step k
plt.plot(solver.x, history[0], label="t=0")
plt.plot(solver.x, history[-1], label="t=end")
plt.legend()
plt.show()
```

```python
solver = PDESolver(dim=2, nx=50, ny=50, Lx=1.0, Ly=1.0)
history = solver.solve(
    equation="diffusion",
    D=0.1,
    dt=0.01,
    steps=200,
    sigma=0.15
)

# Plot initial vs final
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.imshow(history[0].reshape(solver.nx, solver.ny))
plt.title("t=0")
plt.subplot(1,2,2)
plt.imshow(history[-1].reshape(solver.nx, solver.ny))
plt.title("t=end")
plt.show()
```
