# Classical PDE Solver

This notebook shows how to solve PDEs using the Classical solvers implemented with the FiPy library in `ref_solutions.py`. It demonstrates how to use `DiffusionFiPySolver` and `BurgersFiPySolver` classes to solve and visualize solutions to the Diffusion and Burgers' equations in both one and two dimensions.

```python
# imports

import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent))

import matplotlib.pyplot as plt
from src.ref_solutions import DiffusionFiPySolver, BurgersFiPySolver
```

### Burgers 1D

```python
solver = BurgersFiPySolver(dim=1, nx=20, Lx=1.0)
history = solver.solve(
    nu=0.1,
    dt=0.01,
    steps=100,
    sigma=0.1
)

# history[k] = solution at step k
plt.plot(solver.x, history[0], label="t=0", lw=2)
plt.plot(solver.x, history[-1], label="t=end", lw=2)
plt.xlabel("x")
plt.ylabel("f(x,t)")
plt.legend()
plt.show()
```

### Burgers 2D

```python
solver = BurgersFiPySolver(dim=2, nx=20, ny=20, Lx=1.0, Ly=1.0)
history = solver.solve(
    nu=0.1,
    dt=0.01,
    steps=200,
    sigma=0.1
)

# history[k] = solution at step k
plt.subplot(1, 2, 1)
im0 = plt.imshow(history[0].reshape(solver.nx, solver.ny))
plt.title("t=0")
plt.xlabel("y")
plt.ylabel("x")
plt.colorbar(im0, fraction=0.046, pad=0.04, label="f(x, y, t)")

plt.subplot(1, 2, 2)
im1 = plt.imshow(history[-1].reshape(solver.nx, solver.ny))
plt.title("t=end")
plt.xlabel("y")
plt.ylabel("x")
plt.colorbar(im1, fraction=0.046, pad=0.04, label="f(x, y, t)")

plt.tight_layout()
plt.show()
```

### Diffusion 1D

```python
solver = DiffusionFiPySolver(dim=1, nx=20, Lx=1.0)
history = solver.solve(
    D=0.1,
    dt=0.01,
    steps=200,
    sigma=0.15
)

# history[k] = solution at step k
plt.plot(solver.x, history[0], label="t=0", lw=2)
plt.plot(solver.x, history[-1], label="t=end", lw=2)
plt.xlabel("x")
plt.ylabel("f(x,t)")
plt.legend()
plt.show()
```
