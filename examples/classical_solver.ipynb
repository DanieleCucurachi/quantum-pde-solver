```python
solver = BurgersFiPySolver(dim=2, nx=20, ny=20, Lx=1.0, Ly=1.0)
history = solver.solve(
    nu=0.1,
    dt=0.01,
    steps=200,
    sigma=0.1
)

# history[k] = solution at step k
plt.subplot(1,2,1)
plt.imshow(history[0].reshape(solver.nx, solver.ny))
plt.title("t=0")
plt.subplot(1,2,2)
plt.imshow(history[-1].reshape(solver.nx, solver.ny))
plt.title("t=end")
plt.show()
```
